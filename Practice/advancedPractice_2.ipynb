**1. Hastane Otamasyon Sistemi**

**Senaryo**

Bir hastane, hasta kayıtlarını yönetmek için bir otomasyon sistemi geliştirmek istiyor. Bu sistemde, hastaların adları, yaşları ve teşhis bilgileri saklanmalıdır. Hastane, yeni hasta ekleyebilmeyi, mevcut hastaları görüntüleyebilmeyi, belirli bir hastayı sorgulayabilmeyi ve hasta kayıtlarını silebilmeyi istemektedir.

**Sorun**

Hastane, hasta bilgilerini manuel olarak yönetmektedir ve bu durum hatalara yol açmaktadır. Yeni bir hasta ekleme, mevcut hastaları sorgulama ve güncelleme işlemleri zaman alıcı ve verimsiz hale gelmektedir. Bu nedenle, hasta verilerini daha verimli bir şekilde saklamak ve yönetmek için bir yazılım çözümüne ihtiyaç duyulmaktadır.

**Talimatlar**

    1. Yeni Hasta Ekleme:
        Kullanıcı, yeni bir hastanın adını, yaşını ve teşhisini girmelidir.
        Bu bilgiler, hasta kayıt sistemine eklenmelidir.

    2. Hasta Bilgilerini Görüntüleme:
        Kullanıcı, sistemde mevcut tüm hastaların adlarını, yaşlarını ve teşhislerini görüntüleyebilmelidir.

    3. Belirli Bir Hastayı Sorgulama:
        Kullanıcı, belirli bir hastanın adını girerek bu hastanın sistemde olup olmadığını kontrol edebilmelidir.
        Eğer hasta sistemde mevcutsa, hastanın tüm bilgileri görüntülenmelidir.
        Eğer hasta sistemde mevcut değilse, kullanıcıya uygun bir hata mesajı gösterilmelidir.

    4. Hasta Silme:
        Kullanıcı, silmek istediği hastanın adını girmelidir.
        Eğer hasta sistemde mevcutsa, bu hasta sistemden silinmelidir.
        Eğer hasta sistemde mevcut değilse, kullanıcıya uygun bir hata mesajı gösterilmelidir.

    5. Programdan Çıkış:
        Kullanıcı, programdan çıkmak için bir seçenek kullanabilmelidir.

Program Akışı

* Kullanıcı, işlem seçeneklerini menüden seçer.
* Kullanıcı, seçtiği işleme göre gerekli bilgileri girer.
* Program, kullanıcının girdilerine göre gerekli işlemleri yapar ve sonucu kullanıcıya bildirir.
* Program, kullanıcı "Çıkış" seçeneğini seçene kadar çalışmaya devam eder.

**Örnek Kullanıcı Etkileşimi**

```

Hastane Otomasyon Sistemi
1. Yeni Hasta Ekle
2. Hasta Bilgilerini Görüntüle
3. Belirli Bir Hastayı Sorgula
4. Hasta Sil
5. Çıkış

Bir işlem seçin (1-5): 1
Hasta Adı: Ahmet
Hasta Yaşı: 50
Hasta Teşhisi: Kalp Hastalığı
Yeni hasta eklendi: {'ad': 'Ahmet', 'yas': 50, 'teshis': 'Kalp Hastalığı'}

Bir işlem seçin (1-5): 2
Mevcut Hastalar:
1. Hasta - Ad: Ali, Yaş: 45, Teşhis: Grip
2. Hasta - Ad: Ayşe, Yaş: 30, Teşhis: Yüksek Tansiyon
3. Hasta - Ad: Mehmet, Yaş: 60, Teşhis: Diyabet
4. Hasta - Ad: Ahmet, Yaş: 50, Teşhis: Kalp Hastalığı

Bir işlem seçin (1-5): 3
Sorgulanacak hastanın adı: Mehmet
Hasta bulundu: Ad: Mehmet, Yaş: 60, Teşhis: Diyabet

Bir işlem seçin (1-5): 4
Silinecek hastanın adı: Ayşe
Ayşe adlı hasta silindi.

Bir işlem seçin (1-5): 5
Programdan çıkılıyor.
```

````python
# Hastane veri yapıları
hastalar = []

# Hastane otomasyon sistemi menüsü
print("Hastane Otomasyon Sistemi")
print("1. Yeni Hasta Ekle")
print("2. Hasta Bilgilerini Görüntüle")
print("3. Belirli Bir Hastayı Sorgula")
print("4. Hasta Sil")
print("5. Çıkış")

while True:
    secim = input("Bir işlem seçin (1-5): ")

    if secim == "1":
        # Yeni hasta ekleme işlemi
        hasta_adi = input("Hasta Adı: ")
        hasta_yasi = int(input("Hasta Yaşı: "))
        hasta_teshisi = input("Hasta Teşhisi: ")
        yeni_hasta = {"ad": hasta_adi, "yas": hasta_yasi, "teshis": hasta_teshisi}
        hastalar.append(yeni_hasta)
        print(f"Yeni hasta eklendi: {yeni_hasta}")

    elif secim == "2":
        # Hasta bilgilerini görüntüleme işlemi
        if not hastalar:
            print("Kayıtlı hasta bulunmamaktadır.")
        else:
            print("Mevcut Hastalar:")
            for i, hasta in enumerate(hastalar, start=1):
                print(f"{i}. Hasta - Ad: {hasta['ad']}, Yaş: {hasta['yas']}, Teşhis: {hasta['teshis']}")

    elif secim == "3":
        # Belirli bir hastayı sorgulama işlemi
        hasta_adi = input("Sorgulanacak hastanın adı: ")
        bulunan_hasta = None
        for hasta in hastalar:
            if hasta["ad"].lower() == hasta_adi.lower():
                bulunan_hasta = hasta
                break
        if bulunan_hasta:
            print(f"Hasta bulundu: Ad: {bulunan_hasta['ad']}, Yaş: {bulunan_hasta['yas']}, Teşhis: {bulunan_hasta['teshis']}")
        else:
            print(f"{hasta_adi} adlı hasta bulunamadı.")

    elif secim == "4":
        # Hasta silme işlemi
        hasta_adi = input("Silinecek hastanın adı: ")
        silinecek_hasta = None
        for hasta in hastalar:
            if hasta["ad"].lower() == hasta_adi.lower():
                silinecek_hasta = hasta
                break
        if silinecek_hasta:
            hastalar.remove(silinecek_hasta)
            print(f"{silinecek_hasta['ad']} adlı hasta silindi.")
        else:
            print(f"{hasta_adi} adlı hasta bulunamadı.")

    elif secim == "5":
        # Programdan çıkış
        print("Programdan çıkılıyor.")
        break

    else:
        print("Geçersiz bir seçim yaptınız, lütfen 1-5 arasında bir sayı girin.")
````

**2. Satranç Tahtası**

**Senaryo**

Bir satranç eğitmeni, öğrencilerine satranç tahtasındaki taşların konumlarını ve hareketlerini öğretmek için bir yazılım geliştirmek istiyor. Bu yazılım, satranç taşlarının türlerini ve konumlarını saklamalı, belirli bir konumdaki taşı bulabilmeli, bir taşı yeni bir konuma taşıyabilmeli ve taşları tahtadan kaldırabilmelidir. Eğitmen, bu işlemleri basit bir Python programı ile gerçekleştirmek istemektedir.

**Sorun**

Satranç eğitmeni, öğrencilerine taşların hareketlerini ve tahtadaki yerleşimlerini anlatmak için taşların konumlarını manuel olarak takip etmekte zorlanıyor. Ayrıca, belirli taşların yerlerini değiştirmek ve taşları tahtadan kaldırmak da zaman alıcı ve hataya açık hale gelmektedir. Bu nedenle, taşların konumlarını ve hareketlerini daha etkili bir şekilde yönetmek için bir yazılım çözümüne ihtiyaç duyulmaktadır.

**Talimatlar**

    1. Mevcut Durumu Görüntüleme:
        Satranç tahtasındaki tüm taşların türlerini ve konumlarını görüntüleyin.

    2. Belirli Bir Konumdaki Taşı Bulma:
        Kullanıcı, belirli bir konumdaki taşı sorgulayabilmelidir.
        Bu işlem sonucunda, taşın türü ve konumu görüntülenmelidir.

    3. Bir Taşı Yeni Bir Konuma Taşıma:
        Kullanıcı, belirli bir taşı yeni bir konuma taşıyabilmelidir.
        Bu işlem gerçekleştirilirken, taşın eski konumu güncellenmelidir.


    4. Bir Taşı Tahtadan Kaldırma:
        Kullanıcı, belirli bir konumdaki taşı tahtadan kaldırabilmelidir.


**Program Akışı**

* Mevcut satranç tahtasındaki taşların türlerini ve konumlarını içeren tuple'lar oluşturulur.
* Bu tuple'lar bir listeye eklenir.
* Kullanıcı, belirli bir konumdaki taşı sorgular ve taşın türü ve konumu görüntülenir.
* Kullanıcı, belirli bir taşı yeni bir konuma taşır ve taşın konumu güncellenir.
* Kullanıcı, belirli bir konumdaki taşı tahtadan kaldırır ve güncel durum görüntülenir.

**Örnek Kullanıcı Etkileşimi**

```
Satranç Tahtası: [('Piyon', 'a2'), ('At', 'g8'), ('Fil', 'c1'), ('Vezir', 'd1'), ('Kale', 'h1'), ('Şah', 'e1'), ('Piyon', 'h2'), ('Piyon', 'a7')]

'g8' konumundaki taş: ('At', 'g8')

Yeni konuma taşıma: 'g8' konumundaki taş 'f6' konumuna taşındı.
Güncel taş bilgisi: [('Piyon', 'a2'), ('At', 'f6'), ('Fil', 'c1'), ('Vezir', 'd1'), ('Kale', 'h1'), ('Şah', 'e1'), ('Piyon', 'h2'), ('Piyon', 'a7')]

Taşı tahtadan kaldırma: 'c1' konumundaki taş tahtadan kaldırıldı.
Son durum: [('Piyon', 'a2'), ('At', 'f6'), ('Vezir', 'd1'), ('Kale', 'h1'), ('Şah', 'e1'), ('Piyon', 'h2'), ('Piyon', 'a7')]

```

````python
chess_pieces = [
    ('Piyon', 'a2'), 
    ('At', 'g8'), 
    ('Fil', 'c1'), 
    ('Vezir', 'd1'), 
    ('Kale', 'h1'), 
    ('Şah', 'e1'), 
    ('Piyon', 'h2'), 
    ('Piyon', 'a7')
]

def display_board(pieces):
    print("Satranç tahtası:", pieces)
    
    
def find_piece_at_position(pieces, position):
    for piece in pieces:
        if piece[1] == position:
            return piece
    return None


def move_piece(pieces, old_position, new_position):
    for i, piece in enumerate(pieces):
        if piece[1] == old_position:
            pieces[i] = (piece[0], new_position)
            return pieces
    return pieces


def remove_piece(pieces, position):
    for i, piece in enumerate(pieces):
        if piece[1] == position:
            pieces.pop(i)
            return pieces
        
    return pieces


# Mevcut durumu görüntüleme
display_board(chess_pieces)

# Belirli bir konumdaki taşı bulma
position = "g8"
piece = find_piece_at_position(chess_pieces, position)
print(f"'{position}' konumundaki taş: {piece}")


# Bir taşı yeni bir konuma taşıma

old_position = "g8"
new_position = "f6"
chess_pieces = move_piece(chess_pieces, old_position, new_position)

print(f"""
      \nYeni konuma taşıma: 
      '{old_position}' konumundaki taş,
      '{new_position}' konumuna taşınmıştır.
      """)
display_board(chess_pieces)


# Bir taşı tahtadan kaldırma

remove_position = "c1"
chess_pieces = remove_piece(chess_pieces, remove_position)
print(f"""
      Taşı tahtadan kaldırma:
      '{remove_position}' konumundaki taş tahtadan kaldırıldı.
      """)

display_board(chess_pieces)
````

**3. Kurs Seçim Sistemi**

**Senaryo**

Bir üniversitede, öğrencilerin ders seçimlerini yönetmek için bir sistem geliştiriyorsunuz. Öğrenciler, her dönem çeşitli dersler arasından seçim yaparak kendi ders programlarını oluştururlar. Ancak, manuel olarak yapılan ders seçimleri zaman alıcı ve hata yapmaya müsaittir. Bu nedenle, öğrencilerin ders ekleme, ders çıkarma ve mevcut derslerini görüntüleme işlemlerini kolaylaştırmak için bir yazılım çözümü gereklidir.

**Sorun**

Öğrenciler, seçtikleri dersleri elle not almakta ve düzenlemekte zorlanıyorlar. Hangi dersleri seçtiklerini unutmaları veya yanlış not almaları sıkça karşılaşılan sorunlardan biridir. Ayrıca, ders seçimlerini güncellemeleri gerektiğinde bu işlemler oldukça zahmetli ve hata yapmaya açıktır. Bu nedenle, daha verimli ve güvenilir bir sistem gereklidir.

**Talimatlar**

    1. Ders Ekleme:
        Öğrenci, sistem üzerinden yeni bir ders ekleyebilmelidir.
        Eklenen ders, mevcut dersler listesine dahil edilmelidir.

    2. Ders Çıkarma:
        Öğrenci, sistem üzerinden mevcut bir dersi çıkarabilmelidir.
        Çıkarılan ders, mevcut dersler listesinden silinmelidir.

    3. Mevcut Dersleri Görüntüleme:
        Öğrenci, sistem üzerinden mevcut seçtiği dersleri görüntüleyebilmelidir.
        Mevcut dersler ekranda listelenmelidir.

    4. Çıkış:
        Öğrenci, sistemden çıkış yapabilmelidir.

**Program Akışı**

* Öğrencinin seçtiği dersleri tutan bir set oluşturulur.
* Kullanıcıya dört seçenek sunan bir menü görüntülenir:
        Ders Ekle
        Ders Çıkar
        Mevcut Dersleri Görüntüle
        Çıkış
* Kullanıcının seçimine göre ilgili fonksiyon çağrılır ve işlem gerçekleştirilir.
* Ders ekleme veya çıkarma işlemleri set veri yapısı kullanılarak yapılır.
* Mevcut dersler listesi ekrana yazdırılır.
* Kullanıcı çıkış yapmak istediğinde, program sonlandırılır.

**Örnek Kullanıcı Etkileşimi**

```

1. Ders Ekle
2. Ders Çıkar
3. Mevcut Dersleri Görüntüle
4. Çıkış

Seçiminiz (1-4): 1
Eklenecek dersin adını girin: Matematik
Ders eklendi: Matematik

1. Ders Ekle
2. Ders Çıkar
3. Mevcut Dersleri Görüntüle
4. Çıkış

Seçiminiz (1-4): 3
Mevcut dersler: Matematik

1. Ders Ekle
2. Ders Çıkar
3. Mevcut Dersleri Görüntüle
4. Çıkış

Seçiminiz (1-4): 4
Çıkış yapılıyor...
```

````python
courses = set()

def add_course(courses, course):
    courses.add(course)
    print(f"Ders eklendi: {course}")
    
def remove_course(courses, course):
    if course in courses:
        courses.remove(course)
        print(f"Ders çıkarıldı: {course}")
    else:
        print(f"Ders bulunamadı: {course}")
        
def display_courses(courses):
    if len(courses) > 0:
        print(f"Mevcut dersler: {courses}")
    
    else:
        print("Henüz ders seçilmedi.")
        
while True:
    print("""
          1. Ders Ekle
          2. Ders Çıkar
          3. Mevcut Dersleri Görüntüle
          4. Çıkış
          """)
    
    
    choice = input("Seçiminiz (1-4): ")
    
    if choice == "1":
        course = input("Eklenecek dersin adını giriniz: ")
        add_course(courses, course)
        
    elif choice == "2":
        course = input("Çıkarılacak dersin adını giriniz: ")
        remove_course(courses, course)
        
    elif choice == "3":
        display(courses)
        
    elif choice == "4":
        print("Çıkış yapılıyor...")
        break
    
    else:
        print("Geçersiz seçim, lütfen 1-4 arasında bir sayı giriniz.")
````

**4. Sayı Tahmin Oyunu**

**Senaryo**

Bir grup öğrenci, boş zamanlarında oynayabilecekleri eğlenceli bir oyun geliştirmek istiyor. Oyun, basit ve anlaşılır olmalı ve oyunculara eğlenceli bir şekilde vakit geçirme imkanı sunmalıdır. Bu doğrultuda, bir sayı tahmin oyunu geliştirmeye karar veriyorlar. Bu oyun, oyuncuların 1 ile 100 arasında rastgele seçilen bir sayıyı belirli bir deneme sayısı içinde tahmin etmelerini sağlayacak.

**Sorun**

Manuel olarak bir sayı tahmin etme ve kontrol etme işlemleri zaman alıcı ve zorlayıcı olabilir. Ayrıca, oyunun daha organize ve kullanıcı dostu olması için bir yazılım çözümüne ihtiyaç vardır. Bu sayede, kullanıcılar daha rahat bir şekilde tahmin yapabilir ve doğru tahmine ulaşabilirler.

**Talimatlar**

    1. Oyunu Başlatma:
        Oyunculara oyuna hoş geldiniz mesajı verin ve oyunun kurallarını açıklayın.
        1 ile 100 arasında rastgele bir sayı seçin.

    2. Tahmin Yapma:
        Oyuncuların tahmin yapması için 10 deneme hakkı verin.
        Kullanıcıdan geçerli bir sayı girişi alın.
        Kullanıcının tahminini kontrol edin ve doğru olup olmadığını belirtin.

    3. Deneme Hakkı Azaltma:
        Her tahminden sonra kalan deneme hakkını bir azaltın.

    4. Oyunu Kazanma veya Kaybetme:
        Kullanıcı doğru tahminde bulunduğunda oyunu kazandığını belirtin ve oyunu sonlandırın.
        Deneme hakkı bittiğinde kullanıcıya doğru sayıyı gösterin ve oyunu sonlandırın.

    5. Tekrar Oynama Seçeneği:
        Oyuncuya oyunu tekrar oynayıp oynamak istemediğini sorun.
        Kullanıcı tekrar oynamak istiyorsa oyunu yeniden başlatın, istemiyorsa oyunu sonlandırın.

**Program Akışı**

* Oyuncuya oyuna hoş geldiniz mesajı verin ve kuralları açıklayın.
* 1 ile 100 arasında rastgele bir sayı seçin.
* Kullanıcıya 10 deneme hakkı verin ve tahmin yapmasını isteyin.
* Kullanıcının tahminini kontrol edin ve uygun mesajı verin.
* Deneme hakkını bir azaltın.
* Kullanıcı doğru tahminde bulunduğunda oyunu kazandığını belirtin ve oyunu sonlandırın.
* Deneme hakkı bittiğinde doğru sayıyı gösterin ve oyunu sonlandırın.
* Kullanıcıya tekrar oynamak isteyip istemediğini sorun.
* Kullanıcı tekrar oynamak isterse oyunu yeniden başlatın, istemezse oyunu sonlandırın.

**Örnek Kullanıcı Etkileşimi**

```

Sayısal Tahmin Oyununa Hoşgeldiniz!
1 ile 100 arasında bir sayı seçtim. Bakalım tahmin edebilecek misin?

Tahminini gir (Kalan deneme: 10): 50
50 çok yüksek. Daha düşük bir sayı tahmin et.

Tahminini gir (Kalan deneme: 9): 25
25 çok düşük. Daha yüksek bir sayı tahmin et.

Tahminini gir (Kalan deneme: 8): 37
37 doğru tahmin. Tebrikler! Oyunu kazandınız!

Tekrar oynamak ister misiniz? (evet/hayır): evet

```

````python
import random

def start_game():
    print("Sayı Tahmin Oyununa Hoş Geldiniz!")
    print("1 ile 100 arasında bir sayı seçtim. Bakalım tahmin edebilecek misin?")
    
    secret_number = random.randint(1, 100)
    
    attempts = 10
    
    while attempts > 0:
        
        tahmin = input(f"\nTahminini gir (Kalan deneme: {attempts}): ")
        
        if not tahmin.isdigit():
            print("Lütfen geçerli bir sayı giriniz.")
            continue
        
        tahmin = int(tahmin)
        
        if tahmin == secret_number:
            print(f"Tebrikler! {tahmin} doğru tahmin. Oyunu kazandınız!")
            break
            
        elif tahmin < secret_number:
            print(f"{tahmin} çok düşük. Daha yüksek bir sayı tahmin et.")
            
        else:
            print(f"{tahmin} çok yüksek. Daha düşük bir sayı tahmin et.")
            
        attempts -= 1
        
    if attempts == 0:
        print(f"Maalesef deneme hakkınız bitti. Doğru sayı {secret_number} idi.")
            
            
    play_again = input("\nTekrar oynamak ister misiniz? (evet/hayır): ")
        
    if play_again.lower() == "evet":
            start_game()
    else:
        print("Oyun bitti. Teşekkürler")
            
            
start_game()
        
````

**5. Öğrenci Not Sistemi**

**Senaryo**

Bir öğretmen, öğrencilerinin ders notlarını ve durumlarını daha verimli bir şekilde takip edebilmek için bir yazılım geliştirmek istiyor. Bu yazılım, öğrencilerin adlarını ve üç ders için aldıkları notları saklamalı, notların ortalamasını hesaplamalı ve öğrencinin durumunu belirlemelidir. Öğretmen, öğrencilerin durumlarını kolayca görüntüleyebilmek ve gerektiğinde yeni öğrenciler ekleyebilmek istemektedir.

**Sorun**

Manuel olarak öğrencilerin notlarını ve durumlarını takip etmek, hatalara açık ve zaman alıcı olabilir. Bu nedenle, öğrencilerin notlarını saklayacak, ortalamalarını hesaplayacak ve durumlarını belirleyecek bir yazılım çözümüne ihtiyaç vardır. Bu çözüm, öğretmenin işini kolaylaştıracak ve öğrencilerin durumlarını daha doğru bir şekilde değerlendirmesini sağlayacaktır.

**Talimatlar**

    1. Öğrenci Ekleme:
        Kullanıcıdan öğrencinin adını ve üç ders için aldığı notları girmesini isteyin.
        Notların geçerli sayılar olup olmadığını kontrol edin ve geçersiz girişler için uyarı verin.
        Öğrencinin adı ve notları, öğrenci listesine eklenmelidir.

    2. Ortalama Hesaplama:
        Her öğrencinin notlarının ortalamasını hesaplayın.

    3. Durum Belirleme:
        Öğrencinin ortalamasına göre durumunu belirleyin:
            Ortalama 90 ve üzeri ise "Mükemmel"
            Ortalama 75 ve üzeri ise "İyi"
            Ortalama 50 ve üzeri ise "Geçti"
            Ortalama 50'nin altı ise "Kaldı"

    4. Öğrenci Durumlarını Görüntüleme:
        Tüm öğrencilerin adlarını, not ortalamalarını ve durumlarını listeleyin.

    5. Çıkış:
        Kullanıcıya programdan çıkma seçeneği sunun.

**Program Akışı**

* Kullanıcıya ana menüyü görüntüleyin ve seçenekleri sunun.
* Kullanıcı 1'i seçerse, yeni bir öğrenci eklemek için öğrencinin adını ve notlarını alın.
* Kullanıcı 2'yi seçerse, tüm öğrencilerin durumlarını listeleyin.
* Kullanıcı 3'ü seçerse, programdan çıkın.
* Kullanıcı geçersiz bir seçim yaparsa, tekrar menüyü görüntüleyin.

**Örnek Kullanıcı Etkileşimi**

```

1. Öğrenci Ekle
2. Öğrenci Durumlarını Görüntüle
```
