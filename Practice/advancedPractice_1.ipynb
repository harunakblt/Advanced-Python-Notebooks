**1. Mini Personel Yönetim Sistemi**

**Senaryo**

Bir şirket, çalışanlarının bilgilerini yönetmek için bir personel yönetim sistemi geliştirmek istiyor. Bu sistemde her personelin benzersiz bir numarası, adı, departmanı ve maaşı bulunmaktadır. Şirket, yeni personel ekleyebilmeyi, mevcut personeli çıkarabilmeyi, tüm personel listesini görüntüleyebilmeyi ve belirli bir personelin varlığını kontrol edebilmeyi istemektedir.

**Sorun**

Şirket, personel bilgilerini etkin bir şekilde yönetmek için yukarıdaki özellikleri içeren bir program geliştirmekte zorlanmaktadır. Mevcut yöntemleri manuel ve hataya açıktır. Bu nedenle, personel verilerini saklamak ve yönetmek için daha yapılandırılmış ve kullanıcı dostu bir çözüm aramaktadırlar.

**Talimatlar**

    1. Yeni Personel Ekleme:
        Kullanıcı, yeni bir personelin numarasını, adını, departmanını ve maaşını girmelidir.
        Bu bilgiler, personel sistemine eklenmelidir.

    2. Personel Çıkarma:
        Kullanıcı, çıkarmak istediği personelin numarasını girmelidir.
        Eğer personel numarası sistemde mevcutsa, bu personel sistemden çıkarılmalıdır.
        Eğer personel numarası sistemde mevcut değilse, kullanıcıya uygun bir hata mesajı gösterilmelidir.

    3. Personel Listesini Görüntüleme:
        Kullanıcı, sistemde mevcut tüm personelin numaralarını, adlarını, departmanlarını ve maaşlarını görüntüleyebilmelidir.

    4. Personel Varlığını Kontrol Etme:
        Kullanıcı, belirli bir personelin numarasını girerek bu personelin sistemde olup olmadığını kontrol edebilmelidir.
        Eğer personel sistemde mevcutsa, personelin tüm bilgileri görüntülenmelidir.
        Eğer personel sistemde mevcut değilse, kullanıcıya uygun bir hata mesajı gösterilmelidir.

    5. Programdan Çıkış:
        Kullanıcı, programdan çıkmak için bir seçenek kullanabilmelidir.

**Program Akışı**

* Kullanıcı, işlem seçeneklerini menüden seçer.
* Kullanıcı, seçtiği işleme göre gerekli bilgileri girer.
* Program, kullanıcının girdilerine göre gerekli işlemleri yapar ve sonucu kullanıcıya bildirir.
* Program, kullanıcı "Çıkış" seçeneğini seçene kadar çalışmaya devam eder.

**Örnek Kullanıcı Etkileşimi**

```

Personel Yönetim Sistemi
1. Yeni Personel Ekle
2. Personel Çıkar
3. Personel Listesini Görüntüle
4. Personel Varlığını Kontrol Et
5. Çıkış

Bir işlem seçin (1-5): 1
Personel numarası: 104
İsim: Zeynep
Departman: Satış
Maaş: 4800
Yeni personel eklendi: 104 - Zeynep

Bir işlem seçin (1-5): 3
Mevcut Personel Listesi:
Numara: 101, İsim: Ali, Departman: IT, Maaş: 5000
Numara: 102, İsim: Ayşe, Departman: HR, Maaş: 4500
Numara: 103, İsim: Mehmet, Departman: Finance, Maaş: 5500
Numara: 104, İsim: Zeynep, Departman: Satış, Maaş: 4800

Bir işlem seçin (1-5): 4
Kontrol edilecek personel numarası: 105
105 numaralı personel bulunamadı.

Bir işlem seçin (1-5): 5
Programdan çıkılıyor.
```

````python
personel_sistemi = {
    101: {"isim": "Ali", "departman": "IT", "maas": 5000},
    102: {"isim": "Ayşe", "departman": "HR", "maas": 4500},
    103: {"isim": "Mehmet", "departman": "Finance", "maas": 5500},
}

print("""
      Personel Yönetim Sistemi
      1. Yeni Personel Ekle
      2. Personel Çıkar
      3. Personel Listesini Görüntüle
      4. Personel Varlığını Kontrol Et
      5. Çıkış
      """)

while True:
    secim = input("Bir işlem seçiniz (1-5): ")
    
    if secim == "1":
        personel_no = int(input("Personel Numarası: "))
        isim = input("İsim: ")
        departman = input("Departman: ")
        maas = int(input("Maaş: "))
        personel_sistemi[personel_no] = {"isim": isim, "departman": departman, "maas": maas}
        print(f"Yeni personel eklendi: {personel_no} - {isim}")
        
    elif secim == "2":
        personel_no = int(input("Çıkarılacak personel numarası: "))
        if personel_no in personel_sistemi:
            del personel_sistemi[personel_no]
            print(f"Personel çıkarıldı: {personel_no}")
        else:
            print(f"{personel_no} numaralı personel bulunamadı.")
            
    elif secim == "3":
        print("Mevcut Personel Listesi:")
        for personel_no, bilgiler in personel_sistemi.items():
            print(f"Numara: {personel_no} | İsim: {bilgiler['isim']} | Departman: {bilgiler['departman']} | Maaş: {bilgiler['maas']}")
            
    elif secim == "4":
        personel_no = int(input("Kontrol edilecek personel numarası: "))
        if personel_no in personel_sistemi:
            bilgiler = personel_sistemi[personel_no]
            print(f"{personel_no} numaralı personel mevcut. İsim: {bilgiler['isim']}")
        else:
            print(f"{personel_no} numaralı personel bulunamadı.")
        
    elif secim == "5":
        print("Programdan çıkılıyor")
        break
    
    else:
        print("Geçersiz bir seçim yaptınız, lütfen 1-5 arasında bir sayı giriniz.")
````

**2. Hastane Otamasyon Sistemi**

Hastane, hasta bilgilerini manuel olarak yönetmektedir ve bu durum hatalara yol açmaktadır. Yeni bir hasta ekleme, mevcut hastaları sorgulama ve güncelleme işlemleri zaman alıcı ve verimsiz hale gelmektedir. **3. ATM Yazılımı**

**Senaryo**

Bir bankanın müşterileri, banka şubelerine gitmeden hesaplarına erişim sağlamak ve temel bankacılık işlemlerini gerçekleştirmek istemektedir. Banka, müşterilerine bu hizmeti sunabilmek için bir ATM yazılımı geliştirmek istemektedir. Bu yazılım, kullanıcıların hesaplarına giriş yapmalarını, bakiye sorgulamalarını, para yatırma ve çekme işlemlerini gerçekleştirmelerini sağlamalıdır.

**Sorun**

Banka, müşterilerinin hesaplarına hızlı ve güvenli bir şekilde erişim sağlayabilmesi için bir ATM yazılımına ihtiyaç duymaktadır. Mevcut manuel işlemler zaman alıcı ve hata yapmaya açıktır. Ayrıca, müşterilerinin işlemlerini daha rahat ve kolay bir şekilde gerçekleştirebilmesi gerekmektedir. Bu nedenle, kullanıcı dostu bir ATM yazılımı geliştirilmesi gerekmektedir.

**Talimatlar**

    1. Giriş Ekranı:
        Kullanıcı, ATM sistemine giriş yapmak için kullanıcı adı ve şifre girmelidir.
        Eğer kullanıcı adı ve şifre doğruysa, işlemler menüsü görüntülenir.
        Eğer kullanıcı adı veya şifre yanlışsa, hata mesajı gösterilir ve giriş ekranı tekrar gösterilir.

    2. İşlemler Menüsü:
        Kullanıcı giriş yaptıktan sonra aşağıdaki işlemleri gerçekleştirebilir:
            Bakiye Sorgulama
            Para Yatırma
            Para Çekme
            Çıkış

    3. Bakiye Sorgulama:
        Kullanıcı, mevcut bakiyesini görüntüleyebilir.

    4. Para Yatırma:
        Kullanıcı, hesabına para yatırabilir.
        Kullanıcının yatırmak istediği miktar pozitif bir sayı olmalıdır.
        Yatırılan miktar, kullanıcının bakiyesine eklenir.

    5. Para Çekme:
        Kullanıcı, hesabından para çekebilir.
        Kullanıcının çekmek istediği miktar pozitif bir sayı olmalı ve mevcut bakiyesinden fazla olmamalıdır.
        Çekilen miktar, kullanıcının bakiyesinden düşülür.

    6. Çıkış:
        Kullanıcı, işlemler menüsünden çıkış yapabilir.
        Çıkış yapıldığında kullanıcıya veda mesajı gösterilir ve program sona erer.

**Program Akışı**

* Kullanıcı, giriş ekranında kullanıcı adı ve şifre girer.
* Kullanıcı doğrulanırsa, işlemler menüsü görüntülenir.
* Kullanıcı, işlemler menüsünden bir işlem seçer ve ilgili işlemi gerçekleştirir.
* Kullanıcı "Çıkış" seçeneğini seçene kadar işlemler menüsü tekrar gösterilir.

**Örnek Kullanıcı Etkileşimi**

```

ATM Sistemine Hoşgeldiniz
Kullanıcı Adı: user1
Şifre: pass1
Hoşgeldiniz, user1!

İşlemler:
1. Bakiye Sorgulama
2. Para Yatırma
3. Para Çekme
4. Çıkış
Bir işlem seçin (1-4): 1
Mevcut Bakiyeniz: 1000 TL

İşlemler:
1. Bakiye Sorgulama
2. Para Yatırma
3. Para Çekme
4. Çıkış
Bir işlem seçin (1-4): 2
Yatırmak istediğiniz miktar: 500
500.0 TL hesabınıza yatırıldı.

İşlemler:
1. Bakiye Sorgulama
2. Para Yatırma
3. Para Çekme
4. Çıkış
Bir işlem seçin (1-4): 1
Mevcut Bakiyeniz: 1500 TL

İşlemler:
1. Bakiye Sorgulama
2. Para Yatırma
3. Para Çekme
4. Çıkış
Bir işlem seçin (1-4): 4
Çıkış yapılıyor. İyi günler!
```

````python
hesaplar = {
    "user1": {"password": "1234", "balance": 1000},
    "user2": {"password": "2345", "balance": 2000},
    "user3": {"password": "3456", "balance": 3000}
}

print("ATM Sistemine Hoşgeldiniz")
kullanici_adi = input("Kullanıcı Adı: ")
sifre = input("Şifre: ")

if kullanici_adi in hesaplar and hesaplar[kullanici_adi]["password"] == sifre:
    print(f"Hoşgeldiniz, {kullanici_adi}!")
    while True:
        print("""
              İşlemler:
              1. Bakiye Sorgulama
              2. Para Yatırma
              3. Para Çekme
              4. Çıkış
              """)
        
        secim = input("Bir işlem seçiniz (1-4): ")
        
        if secim == "1":
            bakiye = hesaplar[kullanici_adi]["balance"]
            print(f"Mevcut Bakiyeniz: {bakiye} ₺")
            
        elif secim == "2":
            miktar = float(input("Yatırmak istediğiniz miktar: "))
            
            if miktar > 0:
                hesaplar[kullanici_adi]["balance"] += miktar
                print(f"{miktar} ₺ hesabınıza yatırıldı")
                bakiye = hesaplar[kullanici_adi]["balance"]
                print(f"Mevcut Bakiyeniz: {bakiye} ₺")
                
            else:
                print("Girdiğiniz değer geçersiz bir miktrardır.")
                
        elif secim == "3":
            miktar = float(input("Çekmek istediğiniz miktar: "))
            
            if miktar > 0 and miktar <= hesaplar[kullanici_adi]["balance"]:
                hesaplar[kullanici_adi]["balance"] -= miktar
                print(f"{miktar} ₺ hesabınızdan çekildi")
                bakiye = hesaplar[kullanici_adi]["balance"]
                print(f"Mevcut Bakiyeniz: {bakiye} ₺")
                
            else:
                print("Yetersiz bakiye veya geçersiz miktar.")
                
        elif secim == "4":
            print("Çıkış yapılıyor. İyi günler!")
            break
        
        else:
            print("Geçersiz bir seçim yaptınız, lütfen 1-4 arasında bir sayı giriniz.")

else:
    print("Geçersiz kullancı adı yada şifre")
````

**4. Satranç Tahtası**

**Senaryo**

Bir satranç eğitmeni, öğrencilerine satranç tahtasındaki taşların konumlarını ve hareketlerini öğretmek için bir yazılım geliştirmek istiyor. Bu yazılım, satranç taşlarının türlerini ve konumlarını saklamalı, belirli bir konumdaki taşı bulabilmeli, bir taşı yeni bir konuma taşıyabilmeli ve taşları tahtadan kaldırabilmelidir. Eğitmen, bu işlemleri basit bir Python programı ile gerçekleştirmek istemektedir.

**Sorun**

Satranç eğitmeni, öğrencilerine taşların hareketlerini ve tahtadaki yerleşimlerini anlatmak için taşların konumlarını manuel olarak takip etmekte zorlanıyor. Ayrıca, belirli taşların yerlerini değiştirmek ve taşları tahtadan kaldırmak da zaman alıcı ve hataya açık hale gelmektedir. Bu nedenle, taşların konumlarını ve hareketlerini daha etkili bir şekilde yönetmek için bir yazılım çözümüne ihtiyaç duyulmaktadır.

**Talimatlar**

    1. Mevcut Durumu Görüntüleme:
        Satranç tahtasındaki tüm taşların türlerini ve konumlarını görüntüleyin.

    2. Belirli Bir Konumdaki Taşı Bulma:
        Kullanıcı, belirli bir konumdaki taşı sorgulayabilmelidir.
        Bu işlem sonucunda, taşın türü ve konumu görüntülenmelidir.

    3. Bir Taşı Yeni Bir Konuma Taşıma:
        Kullanıcı, belirli bir taşı yeni bir konuma taşıyabilmelidir.
        Bu işlem gerçekleştirilirken, taşın eski konumu güncellenmelidir.
        Döngü ve fonksiyon kullanmadan bu işlemi gerçekleştirin.

    4. Bir Taşı Tahtadan Kaldırma:
        Kullanıcı, belirli bir konumdaki taşı tahtadan kaldırabilmelidir.
        Döngü ve fonksiyon kullanmadan bu işlemi gerçekleştirin.

**Program Akışı**

* Mevcut satranç tahtasındaki taşların türlerini ve konumlarını içeren tuple'lar oluşturulur.
* Bu tuple'lar bir listeye eklenir.
* Kullanıcı, belirli bir konumdaki taşı sorgular ve taşın türü ve konumu görüntülenir.
* Kullanıcı, belirli bir taşı yeni bir konuma taşır ve taşın konumu güncellenir.
* Kullanıcı, belirli bir konumdaki taşı tahtadan kaldırır ve güncel durum görüntülenir.

**Örnek Kullanıcı Etkileşimi**

```
Satranç Tahtası: [('Piyon', 'a2'), ('At', 'g8'), ('Fil', 'c1'), ('Vezir', 'd1'), ('Kale', 'h1'), ('Şah', 'e1'), ('Piyon', 'h2'), ('Piyon', 'a7')]

'g8' konumundaki taş: ('At', 'g8')

Taşı tahtadan kaldırma: 'c1' konumundaki taş tahtadan kaldırıldı.

```
