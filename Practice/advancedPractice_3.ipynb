````python
# Başlangıçta boş gelir ve gider listeleri oluştur
income_list = []
expense_list = []

# Listeye gelir ekleme (manuel olarak birkaç ekleme yapalım)
income_list.append(["Maaş", 5000])
income_list.append(["Freelance Proje", 1500])

# Listeye gider ekleme (manuel olarak birkaç ekleme yapalım)
expense_list.append(["Kira", 1000])
expense_list.append(["Market Alışverişi", 600])
expense_list.append(["Fatura", 300])

# Gelir ve giderleri görüntüleme (manuel olarak yazdırma)
print("Gelir Listesi:")
print("1. Kaynak: " + income_list[0][0] + ", Miktar: " + str(income_list[0][1]) + " TL")
print("2. Kaynak: " + income_list[1][0] + ", Miktar: " + str(income_list[1][1]) + " TL")

print("\nGider Listesi:")
print("1. Harcama: " + expense_list[0][0] + ", Miktar: " + str(expense_list[0][1]) + " TL")
print("2. Harcama: " + expense_list[1][0] + ", Miktar: " + str(expense_list[1][1]) + " TL")
print("3. Harcama: " + expense_list[2][0] + ", Miktar: " + str(expense_list[2][1]) + " TL")

# Bütçe hesaplamaları yapma
total_income = income_list[0][1] + income_list[1][1]
total_expenses = expense_list[0][1] + expense_list[1][1] + expense_list[2][1]
net_budget = total_income - total_expenses

# Bütçe durumunu yazdırma
print("\nBütçe Durumu:")
print("Toplam Gelir: " + str(total_income) + " TL")
print("Toplam Gider: " + str(total_expenses) + " TL")
print("Net Bütçe: " + str(net_budget) + " TL")

# Harcama ekleme (örneğin, ulaşım gideri ekleyelim)
expense_list.append(["Ulaşım", 150])

# Güncellenmiş gider listesini görüntüleme (manuel olarak yazdırma)
print("\nGüncellenmiş Gider Listesi:")
print("1. Harcama: " + expense_list[0][0] + ", Miktar: " + str(expense_list[0][1]) + " TL")
print("2. Harcama: " + expense_list[1][0] + ", Miktar: " + str(expense_list[1][1]) + " TL")
print("3. Harcama: " + expense_list[2][0] + ", Miktar: " + str(expense_list[2][1]) + " TL")
print("4. Harcama: " + expense_list[3][0] + ", Miktar: " + str(expense_list[3][1]) + " TL")

# Güncellenmiş bütçe hesaplamaları yapma
total_expenses = expense_list[0][1] + expense_list[1][1] + expense_list[2][1] + expense_list[3][1]
net_budget = total_income - total_expenses

# Güncellenmiş bütçe durumunu yazdırma
print("\nGüncellenmiş Bütçe Durumu:")
print("Toplam Gelir: " + str(total_income) + " TL")
print("Toplam Gider: " + str(total_expenses) + " TL")
print("Net Bütçe: " + str(net_budget) + " TL")
````

**3. Metin Şifreleme**

Sorun Bildirimi: Bir metni belirli bir şifreleme algoritması kullanarak şifreleyen bir program yazın. Kullanıcı, bir metin ve şifreleme anahtarı sağlayacak ve uygulamanın metni şifrelemesini sağlamalıdır.

Görev: Kullanıcının sağladığı metni belirli bir şifreleme algoritması kullanarak şifreleyin. Şifreleme algoritması olarak, her harfi belirli bir sayı ile kaydırma (Caesar şifrelemesi), harfleri tersine çevirme veya basit bir yer değiştirme yöntemi kullanabilirsiniz.

Bonus: Kullanıcıya şifrelenmiş metni şifresini çözmek için bir seçenek sunun.

````python
metin = input("Şifrelemek istediğiniz metni giriniz:")
anahtar = int(input("Şifreleme anahtarını giriniz (Bir tam sayı):"))


sifrelenmis_metin = ""

for harf in metin:
    if harf.isalpha():
        if harf.isupper():
            sifrelenmis_metin += chr((ord(harf) + anahtar - 65) % 26 + 65)
        
        else:
            sifrelenmis_metin += chr((ord(harf) + anahtar - 97) % 26 + 97)
    
    else:
        sifrelenmis_metin += harf
        
print("Şifrelenmiş Metin:", sifrelenmis_metin)
````

**4.	Benzersiz Karakter Sayacı**

Sorun: Verilen bir metindeki benzersiz karakter sayısını bulun.

Görev: Bir metin girdisi alın, metindeki benzersiz karakterlerin sayısını hesaplayın ve sonucu kullanıcıya gösterin.

Örnek: "python programlama" -> 11 (boşlukları saymaz)


````python
metin = input("Lütfen bir metin giriniz:")

benzersiz_karakterler = set()

for karakter in metin:
    if karakter != " ":
        benzersiz_karakterler.add(karakter)
        

benzersiz_karakterler_sayisi = len(benzersiz_karakterler)

print(f"""Metinde yer alan benzersiz karakterler: {benzersiz_karakterler}, 
      \nbenzersiz karakter sayısı: {benzersiz_karakterler_sayisi}""")
````

**5. Kurs Seçim Sistemi:**

Bir üniversitede öğrencilerin ders seçimlerini yönetmek için bir sistem geliştiriyorsunuz. Her öğrencinin seçtiği dersler set olarak tutulmalıdır. Set kullanarak, bir öğrencinin ders seçim listesine ders eklemesini veya çıkarmasını sağlayan bir program yazın. Ayrıca, bir öğrencinin mevcut ders listesini görüntülemek için bir seçenek ekleyin.



````python
ogrenciler_dersler = {
    "Ali": {"Matematik", "Fizik"},
    "Ayşe": {"Kimya", "Tarih"},
    "Mehmet": {"Tarih", "Coğrafya"}
}

print("""
      Öğrenci Ders Seçim Sistemi
      \n1. Ders Ekle
      \n2. Ders Çıkar
      \n3. Ders Listesini Görüntüle
      \n4. Çıkış
      """)

while True:
    secim = input("Bir işlem seçin(1-4):")
    
    if secim == "1":
        ogrenci_adi = input("Öğrencinin adı: ")
        ders = input("Eklenecek ders: ")
        
        if ogrenci_adi in ogrenciler_dersler:
            ogrenciler_dersler[ogrenci_adi].add(ders)
        else:
            ogrenciler_dersler[ogrenci_adi] = {ders}
            
        print(f"{ogrenci_adi} adlı öğrencinin güncel ders listesi: {ogrenciler_dersler[ogrenci_adi]}")
        
    elif secim == "2":
        ogrenci_adi = input("Öğrencinin adı: ")
        ders = input("Çıkarılacak ders: ")
        
        if ogrenci_adi in ogrenciler_dersler and ders in ogrenciler_dersler[ogrenci_adi]:
            ogrenciler_dersler[ogrenci_adi].discard(ders)
        
            print(f"{ogrenci_adi} adlı öğrencinin güncel ders listesi: {ogrenciler_dersler[ogrenci_adi]}") 
            
        else:
            print(f"{ogrenci_adi} adlı öğrencinin ders listesinde {ders} dersi bulunamadı.")  
            
    elif secim == "3":
        ogrenci_adi = input("Öğrencinin adı: ")
        
        if ogrenci_adi in ogrenciler_dersler:
            print(f"{ogrenci_adi} adlı öğrencinin ders listesi: {ogrenciler_dersler[ogrenci_adi]}") 
            
        else:
            print(f"{ogrenci_adi} adlı öğrenci bulunamadı.")
        
    elif secim == "4":
        print("Programdan çıkılıyor.")
        break
    
    else:
        print("Geçersiz bir seçim yaptınız, lütfen 1-4 arasında bir sayı seçiniz.")
````

**6. Kitap Kütüphanesi Yönetimi:**

Bir kitap kütüphanesinin envanterini set kullanarak yönetmek istiyorsunuz. Her kitabın adı, yazarı ve yayın yılı bulunmaktadır. Kütüphaneye yeni kitap eklemek, mevcut kitapları kontrol etmek ve bir kitabın varlığını adıyla hızlıca kontrol etmek için bir program yazın.

````python
kutuphane = {
    ("1984", "George Orwell", 1949),
    ("To Kill a Mockingbird", "Harper Lee", 1960),
    ("İnce Memed", "Yaşar Kemal", 1967)
}

print("""
      Kütüphane Yönetim Sistemi
      \n1. Yeni Kitap Ekle
      \n2. Mevcut Kitapları Görüntüle
      \n3. Bir Kitabın varlığını Kontrol Et
      \n4. Çıkış
      """)

while True:
    secim = input("Bir işlem seçin (1-4): ")
    
    if secim == "1":
        kitap_adi = input("Kitabın adı: ")
        yazar = input("Yazar: ")
        yayin_yili = int(input("Yayin yılı: "))
        yeni_kitap = (kitap_adi, yazar, yayin_yili)
        kutuphane.add(yeni_kitap)
        print(f"Yeni kitap eklendi: {yeni_kitap}")
        
    elif secim == "2":
        print("Mevcut kitaplar")
        for kitap in kutuphane:
            print(kitap)
            
            
    elif secim == "3":
        kitap_adi = input("Kontrol edilecek kitabın adı: ")
        kitap_var = False
        
        for kitap in kutuphane:
            if kitap[0] == kitap_adi:
                kitap_var = True
                break
            
        if kitap_var:
            print(f"{kitap_adi} kütüphanede mevcut. ")
        else:
            print(f"{kitap_adi} kütüphanede bulanamadı. ")
            
            
    elif secim == "4":
        print("Programdan çıkılıyor.")
        break
    
    else:
        print("Geçersiz bir seçim yaptınız, lütfen 1-4 arasında bir sayı seçiniz.")
````

**7. Mini Personel Yönetim Sistemi**

Senaryo

Bir şirket, çalışanlarının bilgilerini yönetmek için bir personel yönetim sistemi geliştirmek istiyor. Bu sistemde her personelin benzersiz bir numarası, adı, departmanı ve maaşı bulunmaktadır. Şirket, yeni personel ekleyebilmeyi, mevcut personeli çıkarabilmeyi, tüm personel listesini görüntüleyebilmeyi ve belirli bir personelin varlığını kontrol edebilmeyi istemektedir.
Sorun

Şirket, personel bilgilerini etkin bir şekilde yönetmek için yukarıdaki özellikleri içeren bir program geliştirmekte zorlanmaktadır. Mevcut yöntemleri manuel ve hataya açıktır. Bu nedenle, personel verilerini saklamak ve yönetmek için daha yapılandırılmış ve kullanıcı dostu bir çözüm aramaktadırlar.
Talimatlar

    Yeni Personel Ekleme:
        Kullanıcı, yeni bir personelin numarasını, adını, departmanını ve maaşını girmelidir.
        Bu bilgiler, personel sistemine eklenmelidir.

    Personel Çıkarma:
        Kullanıcı, çıkarmak istediği personelin numarasını girmelidir.
        Eğer personel numarası sistemde mevcutsa, bu personel sistemden çıkarılmalıdır.
        Eğer personel numarası sistemde mevcut değilse, kullanıcıya uygun bir hata mesajı gösterilmelidir.

    Personel Listesini Görüntüleme:
        Kullanıcı, sistemde mevcut tüm personelin numaralarını, adlarını, departmanlarını ve maaşlarını görüntüleyebilmelidir.

    Personel Varlığını Kontrol Etme:
        Kullanıcı, belirli bir personelin numarasını girerek bu personelin sistemde olup olmadığını kontrol edebilmelidir.
        Eğer personel sistemde mevcutsa, personelin tüm bilgileri görüntülenmelidir.
        Eğer personel sistemde mevcut değilse, kullanıcıya uygun bir hata mesajı gösterilmelidir.

    Programdan Çıkış:
        Kullanıcı, programdan çıkmak için bir seçenek kullanabilmelidir.

Program Akışı

    Kullanıcı, işlem seçeneklerini menüden seçer.
    Kullanıcı, seçtiği işleme göre gerekli bilgileri girer.
    Program, kullanıcının girdilerine göre gerekli işlemleri yapar ve sonucu kullanıcıya bildirir.
    Program, kullanıcı "Çıkış" seçeneğini seçene kadar çalışmaya devam eder.
