**1. Çalışan Maaş Yönetim Sistemi** ***----ÖDEV----***

Bir şirketin çalışanlarının adlarını ve maaşlarını yönetmek için bir program yazıyorsunuz. Program, kullanıcıya çalışan eklemeyi, mevcut çalışanları listelemeyi, belirli bir maaştan yüksek olan çalışanları filtrelemeyi ve programdan çıkmayı seçenek olarak sunuyor.

**Talimatlar:**

1. Çalışan Ekle:
* Kullanıcıdan yeni bir çalışanın adını ve maaşını girmesi istenir.
* Girilen maaşın geçerli bir sayı olup olmadığı kontrol edilir.
* Geçerli ise, çalışan listesine adı ve maaşıyla birlikte eklenir ve eklenen çalışanın bilgisi kullanıcıya gösterilir.
* Geçersiz maaş girişi durumunda kullanıcıya uyarı verilir ve işlem tekrarlanır.

2. Çalışanları Listele:
* Eğer henüz hiç çalışan eklenmemişse, "Henüz çalışan eklenmedi." mesajı gösterilir.
* Eğer çalışanlar varsa, her çalışanın adı ve maaşı sırasıyla listelenir.

3. Belirli Maaştan Yüksek Olan Çalışanları Filtrele:
* Kullanıcıdan bir minimum maaş değeri girmesi istenir.
* Girilen değerin geçerli bir sayı olup olmadığı kontrol edilir.
* Geçerli ise, girilen maaştan yüksek olan tüm çalışanlar listelenir.
* Bu çalışanlar varsa, her birinin adı ve maaşı gösterilir.
* Belirtilen maaştan yüksek çalışan yoksa, kullanıcıya bilgi verilir.

4. Çıkış:
* Kullanıcı programdan çıkmak istediğinde seçenek sunulur.
* "4" tuşuna basılarak çıkış yapılır ve program sonlandırılır.

````python
employees = []

def add_employee():
    name = input("Çalışan adını giriniz: ")
    salary = input("Maaşını giriniz :")
    if salary.isdigit():
        employees.append({"name": name, "salary" : int(salary)})
        print(f"{name} adlı çalışan eklendi")

    else:
        print("Geçersiz maaş girdiniz.Lütfen tekrar deneyiniz.")

def list_employees():
    if not employees:
        print("Henüz çalışan eklenmedi")
    else:
        for employee in employees:
            print(f"çalışan :{employee['name']}, Maaş : {employee['salary']} ")

def filter_employees_by_salary(min_salary):
    filtered_employees = list(filter(lambda employee: employee['salary'] > min_salary, employees))
    if not filtered_employees:
        print(f"{min_salary} maaşından yüksek maaşta çalışan yok.")
    else:
        for employee in filtered_employees:
            print(f"Çalışan : {employee['name']}, Maaş : {employee['salary']}")

def main():
    while True:
        print("""
              1. Çalışan Ekle
              2. Çalışanları Listele
              3. Belirli Maaştan Yüksek Olan Çalışanları Filtre
              4. Çıkış

""")
        
        choice = input("Lütfen size belirtilen listeden seçim yapınız (1-4):")
        if choice == "1":
            add_employee()
        elif choice == "2":
            list_employees()
        elif choice == "3":
            min_salary = input("Lütfen minimum maaşı giriniz:")
            if min_salary.isdigit():
                filter_employees_by_salary(int(min_salary))
            else:
                print("Geçersiz maaş girdiniz. Lütfen tekrar deneyiniz.")
        elif choice == "4":
            print("Programdan çıkış yapılıyor")
            break
        else:
            print("Geçersiz seçim. Lütfen tekrar deneyiniz.")

if __name__ == "__main__":
    main()

````

**2. Kişisel Finans Yönetim Uygulaması**

**Senaryo:**

Merve, her ay gelirlerini ve giderlerini takip etmekte zorlanıyor. Aylık bütçesini düzenli bir şekilde takip ederek nereye ne kadar harcadığını ve ne kadar birikim yapabileceğini bilmek istiyor. Bu nedenle, gelir ve giderlerini kolayca kaydedebileceği, bütçesini hesaplayabileceği ve bu verileri saklayabileceği bir uygulamaya ihtiyaç duyuyor. Python kullanarak bir "Kişisel Finans Yönetim Uygulaması" geliştirilmesi için sizden yardım istiyor.

**Problem:**

Merve'nin gelir ve giderlerini takip edebileceği, gelir ve gider listelerini görüntüleyebileceği, aylık bütçesini hesaplayabileceği ve bu verileri dosyaya kaydedip daha sonra okuyabileceği bir Python uygulaması geliştirin.

**Talimatlar:**

    1. Gelir Ekleme:
        Kullanıcıdan eklemek istediği geliri alarak gelir listesine ekleyen bir fonksiyon yazın.
        Gelirler "incomes.txt" dosyasına kaydedilecek.

    2. Gider Ekleme:
        Kullanıcıdan eklemek istediği gideri alarak gider listesine ekleyen bir fonksiyon yazın.
        Giderler "expenses.txt" dosyasına kaydedilecek.

    3. Gelir ve Gider Listesi Gösterme:
        Kullanıcının eklediği tüm gelirleri listeleyen bir fonksiyon yazın.
        Kullanıcının eklediği tüm giderleri listeleyen bir fonksiyon yazın.

    4. Aylık Bütçe Hesaplama:
        Gelirler ve giderler arasındaki farkı hesaplayarak kullanıcıya gösteren bir fonksiyon yazın.

    5. Verileri Dosyaya Kaydetme ve Dosyadan Okuma:
        Gelir ve gider verilerini dosyaya kaydeden bir fonksiyon yazın.
        Gelir ve gider verilerini dosyadan okuyarak uygulama başlatıldığında bu verilerin yüklenmesini sağlayan bir fonksiyon yazın.

**Kullanım:**

    1. Gelir Ekleme:
        Uygulamayı başlatın ve "1" seçeneğini seçerek gelir ekleyin.
        Geliri ekledikten sonra, "incomes.txt" dosyasına kaydedilecektir.

    2. Gider Ekleme:
        Uygulamayı başlatın ve "2" seçeneğini seçerek gider ekleyin.
        Gideri ekledikten sonra, "expenses.txt" dosyasına kaydedilecektir.

    3. Gelir ve Gider Listesi Gösterme:
        Gelirleri görmek için "3" seçeneğini seçin.
        Giderleri görmek için "4" seçeneğini seçin.

    4. Aylık Bütçe Hesaplama:
        Gelirler ve giderler arasındaki farkı hesaplamak için "5" seçeneğini seçin.

    5. Verileri Dosyaya Kaydetme ve Dosyadan Okuma:
        Uygulamayı kapatmadan önce verileri kaydetmek için "6" seçeneğini seçin.
        Uygulama her başlatıldığında "incomes.txt" ve "expenses.txt" dosyalarından verileri okuyarak listeyi günceller.

**Örnek Kullanım:**

1. Başlatma:

```
python personal_finance.py
```

2. Gelir Ekleme:

```

```

3. Gider Ekleme:

```

```

4. Gelirleri Gösterme:

```

- 5000
```

5. Giderleri Gösterme:

```

Gider Listesi:
- 2000
```

6. Aylık Bütçeyi Hesaplama:

```

```

7. Verileri Kaydetme ve Çıkma:

```
Bir seçenek seçin (1-6): 6
Veriler kaydedildi. Çıkılıyor...
```

````python
import os

def add_income(income_list):
    income = float(input("Eklemek istediğiniz geliri giriniz: "))
    income_list.append(income)
    
def add_expense(expense_list):
    expense = float(input("Ekelemek istediğniz gideri giriniz: "))
    expense_list.append(expense)
    
def show_incomes(income_list):
    print("Gelir Listesi: ")
    for income in income_list:
        print(f"- {income}")
        
def show_expenses(expense_list):
    print("Gider Listesi: ")
    for expense in expense_list:
        print(f"- {expense}")
        
        
def calculate_budget(income_list, expense_list):
    total_income = sum(income_list)
    total_expense = sum(expense_list)
    budget = total_income - total_expense
    return budget

def save_to_file(filename, data):
    with open(filename, "w") as file:
        for item in data:
            file.write(f"{item}\n")
            
def read_from_file(filename):
    if not os.path.exists(filename):
        return []
    with open(filename, "r") as file:
        return [float(line.strip()) for line in file.readlines()]
    
    
def main():
    income_list = read_from_file("incomes.txt")
    expense_list = read_from_file("expenses.txt")
    
    while True:
        print("""
              Kişisel Finans Yönetim Uygulaması
              ---------------------------------
              1. Gelir Ekle
              2. Gider Ekle
              3. Gelirleri Göster
              4. Giderleri Göster
              5. Aylık Bütçeyi Hesapla
              6. Çık ve verileri kaydet
              """)
        
        
        choice = input("Bir seçenek seçiniz (1-6): ")
        
        if choice == "1":
            add_income(income_list)
            
        elif choice == "2":
            add_expense(expense_list)
            
        elif choice == "3":
            show_incomes(income_list)
            
        elif choice == "4":
            show_expenses(expense_list)
            
        elif choice == "5":
            budget = calculate_budget(income_list, expense_list)
            print(f"Aylık bütçeniz: {budget}")
            
        elif choice == "6":
            save_to_file("incomes.txt", income_list)
            save_to_file("expenses.txt", expense_list)
            print("Veriler kaydedildi. Çıkış yapılıyor...")
            break
        
        else:
            print("Geçersiz seçenek. Lütfen tekrar deneyiniz.")
            

if __name__ == "__main__":
    main()
````

**4. Taş, Kağıt, Makas Oyunu**

**Senaryo:**

Bir kullanıcı ve bilgisayar arasında oynanan basit bir "Taş, Kağıt, Makas" oyunu geliştirmek istiyorsunuz. Oyunun amacı, kullanıcının taş, kağıt veya makas seçeneklerinden birini seçerek bilgisayarla rekabet etmesidir. Kullanıcı ve bilgisayarın seçimine göre oyun sonucunu belirlemek ve kullanıcının tekrar oynamak isteyip istemediğini sormak gerekmektedir.

**Amaç:**

* Kullanıcıdan bir seçim almak (taş, kağıt veya makas).
* Bilgisayarın rastgele bir seçim yapmasını sağlamak.
* Kullanıcı ve bilgisayarın seçimlerini karşılaştırarak kazananı belirlemek.
* Sonucu ekrana yazdırmak.
* Kullanıcının tekrar oynamak isteyip istemediğini sormak ve buna göre oyunu tekrar başlatmak veya sonlandırmak.

**Talimatlar:**

 * Kullanıcı Seçimi: Program, kullanıcıdan taş, kağıt veya makas seçeneklerinden birini girmesini ister. Geçerli bir seçim yapılana kadar kullanıcıdan tekrar tekrar giriş yapması istenir.

* Bilgisayar Seçimi: Bilgisayar, taş, kağıt veya makas seçeneklerinden birini rastgele seçer.

* Sonuç Belirleme: Kullanıcının ve bilgisayarın seçimleri karşılaştırılır ve sonuç belirlenir:
        Taş makası yener.
        Kağıt taşı yener.
        Makas kağıdı yener.
        Aynı seçim yapılırsa sonuç berabere olur.

* Sonuç Görüntüleme: Kullanıcının ve bilgisayarın seçimleri ekrana yazdırılır ve sonuç (kazandınız, kaybettiniz veya berabere) kullanıcıya bildirilir.

* Tekrar Oynama: Kullanıcıya tekrar oynamak isteyip istemediği sorulur. Kullanıcı "evet" derse oyun yeniden başlar, "hayır" derse oyun sonlanır.

**Kodu Çalıştırmak İçin Adımlar:**

*Python Yüklü Olmalı: Bilgisayarınızda Python'un yüklü olduğundan emin olun.
* Terminal veya Komut Satırını Açın: Terminal veya komut satırını açın ve kod dosyasının bulunduğu dizine gidin.
* Kodu Çalıştırın: Aşağıdaki komutu terminal veya komut satırına girerek kodu çalıştırın:

```
python tas_kagit_makas.py
```

* Talimatları Takip Edin: Programın talimatlarını takip edin ve oyun oynayın.

**Örnek Kullanım:**

```
Taş, Kağıt, Makas Oyununa Hoşgeldiniz!
Seçiminizi yapın (taş, kağıt, makas): taş
Siz: taş - Bilgisayar: makas
Kazandınız!
Tekrar oynamak ister misiniz? (evet/hayır): evet
Seçiminizi yapın (taş, kağıt, makas): kağıt
Siz: kağıt - Bilgisayar: taş
Kazandınız!
Tekrar oynamak ister misiniz? (evet/hayır): hayır
```

````python
import random

materials = ["taş","kağıt","makas"]

def taş_kağıt_makas(user_choice):
    pc_choice = random.choice(materials)
    if user_choice == pc_choice:
        print("Berabere")
    elif user_choice == pc_choice :
        print(f" Pc seçimi : {pc_choice} , Sen Kazandın")
    elif user_choice == "taş" and pc_choice == "kağıt":
        print(f" Pc seçimi : {pc_choice} , Sen Kaybettin")
    elif user_choice == "taş" and pc_choice == "makas":
        print(f" Pc seçimi : {pc_choice} , Sen Kazandın")
    elif user_choice == "kağıt" and pc_choice == "taş":
        print(f" Pc seçimi : {pc_choice} , Sen Kazandın")
    elif user_choice == "kağıt" and pc_choice == "makas":
        print(f" Pc seçimi : {pc_choice} , Sen Kaybettin")
    elif user_choice == "makas" and pc_choice == "kağıt":
        print(f" Pc seçimi : {pc_choice} , Sen Kazandın")
    elif user_choice == "makas" and pc_choice == "taş":
        print(f" Pc seçimi : {pc_choice} , Sen Kaybettin")
    else:
        print("Lütfen Taş, Kağıt veya Makas yazınız ")
        taş_kağıt_makas(user_choice)

def main():
    while True:
        user_choice = input("Taş, Kağıt veya Makas, Birini Seç : ")
        taş_kağıt_makas(user_choice)
        devam = input("Tekrar Oynamak İçin 'E' , Çıkmak İçin 'q' yazınız:")
        if devam == "q":
            print("Çıkış yapılıyor. teşekkürler.")
            break
        elif devam == "E".lower():
            continue


if __name__ == "__main__":
    main()      
````

**5. Görev Yönetim Sistemi**

Bir grup arkadaş, birlikte çeşitli etkinlikler ve projeler planlamak ve takip etmek için basit bir görev yönetim sistemi kullanmak istiyor. Bu sistem, her arkadaşın görev eklemesine, mevcut görevleri listelemesine, görevleri tamamlanmış olarak işaretlemesine ve tamamlanmış görevleri temizlemesine olanak tanır.

**Talimatlar:**

**Amaç:**

Bu görev yönetim sistemi, kullanıcıların görev eklemesini, görevleri listelemesini, tamamlanmış görevleri işaretlemesini ve tamamlanmış görevleri temizlemesini sağlar.

**Adım 1: Gereksinimler**

**Adım 4: Kullanım Talimatları**

    Görev Ekle: Menüden 1'i seçin ve eklemek istediğiniz görevin tanımını girin.
    Görevleri Listele: Menüden 2'yi seçin ve mevcut görevleri görün.
    Görevi Tamamlandı Olarak İşaretle: Menüden 3'ü seçin ve tamamlanmış olarak işaretlemek istediğiniz görevin numarasını girin.
    Tamamlanmış Görevleri Temizle: Menüden 4'ü seçin ve tamamlanmış görevler listeden silinsin.
    Çık: Menüden 5'i seçin ve programdan çıkın.

````python
import random

def derece_yukseltme(derece):
    if derece == 1:
        return (2, 25, 0)
    elif derece == 2:
        return (3, 20, 1)
    elif derece == 3:
        return (4, 15, 1)
    elif derece == 4:
        return (5, 10, 1)
    else:
        return (0, 0, 0)  # Yanlış dereceler için

def item_yukseltme_sansli(derece):
    if derece == 1:
        sans = 25
    elif derece == 2:
        sans = 20
    elif derece == 3:
        sans = 15
    elif derece == 4:
        sans = 10
    else:
        sans = 0

    rastgele_sayi = random.randint(1, 100)
    return rastgele_sayi <= sans

def main():
    derece = 1

    while True:
        if derece == 0:
            print("Item yok oldu. Oyun bitti.")
            break
        elif derece == 5:
            print("Item derecesi 5 oldu. Oyunu kazandınız!")
            break
        
        print(f"Şu anki item derecesi: {derece}")
        devam = input("Dereceyi yükseltmek ister misiniz? (e/h): ").lower()
        
        if devam != 'e':
            print("Oyun sonlandırıldı.")
            break
        
        if item_yukseltme_sansli(derece):
            derece, _, _ = derece_yukseltme(derece)
            print(f"Tebrikler! Item'in derecesi {derece} oldu.")
        else:
            _, _, derece_dusus = derece_yukseltme(derece)
            derece -= derece_dusus
            print(f"Başarısız! Item'in derecesi {derece} oldu.")

if __name__ == "__main__":
    main()
````
